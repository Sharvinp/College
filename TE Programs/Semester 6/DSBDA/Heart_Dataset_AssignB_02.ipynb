```
d=d.replace(to_replace='No',value=0)
d=d.replace(to_replace='Yes',value=1)
d
```

```
d=d.replace(to_replace='fixed',value=1)
d=d.replace(to_replace='normal',value=2)
d=d.replace(to_replace='reversable',value=3)
```

```
mean_data=d["Thal"].mean()
print(mean_data)
mean_data1=d['Ca'].mean()
print(mean_data1)
```

```
d['Thal'].fillna(value=mean_data,inplace=True)
d['Ca'].fillna(value=mean_data1,inplace=True)
d.isnull().sum()
```

```
#train test model
X=d[['Age','Sex','RestBP','Chol','Fbs','Ca','Thal']]
Y=d[['AHD']]
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.3, random_state=100)
```

```
#Decision Tree (entropy)

from sklearn.tree import DecisionTreeClassifier
e = DecisionTreeClassifier(criterion='entropy', max_depth=5,min_samples_leaf=4)
e.fit(x_train,y_train)
```

```
#Decision Tree (Gini)
g= DecisionTreeClassifier(criterion='gini',max_depth=5,min_samples_leaf=4)
g.fit(x_train,y_train)
```

```
#x test prediction -gini
g_pred= g.predict(x_test)
g_pred
```

```
#x train predicition -gini
g_pred1= g.predict(x_train)
g_pred1
```

```
#x train prediction -entropy
e_pred= e.predict(x_train)
e_pred
```

```
#x test prediction -entropy
e_pred1= e.predict(x_test)
e_pred1
```

```
#train accuracy entropy 
from sklearn import metrics
metrics.accuracy_score(y_train,e_pred)
```

```
#test accuracy entropy
metrics.accuracy_score(y_test,e_pred1)
```

```
#train accuracy gini 
metrics.accuracy_score(y_train,g_pred1)
```

```
#test accuracy gini
metrics.accuracy_score(y_test,g_pred)
```

```
#confusion matrix test -entropy
metrics.confusion_matrix(y_test ,e_pred1)
```

```
#confusion matrix train - entropy
metrics.confusion_matrix(y_train ,e_pred)
```

```
#confusion matrix train - gini
metrics.confusion_matrix(y_train ,g_pred1)
```

```
#confusion matrix test-gini
metrics.confusion_matrix(y_test ,g_pred)
```

```
#gini tree plot
from sklearn import tree
tree.plot_tree(g)
```
